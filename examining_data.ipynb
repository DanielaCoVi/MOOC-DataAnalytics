### Rellenar la columna "course_id" con el mismo dato que existe en esa misma columna si coincide el "reviewer" que da el "review" de esa fila
#### Esta lógica crea una lista de "reviewers" y otra lista de "course_ids". Luego, para cada fila con un valor "NaN" en la columna "course_id", busca hacia abajo en las siguientes filas hasta encontrar una que tenga el mismo "reviewer" y un valor válido en "course_id". Si se encuentra, se rellena la fila con ese valor y se sale del loop interno. Finalmente, se reemplazan los valores en la columna "course_id" del DataFrame con la lista actualizada

```python
reviewers_dict = {}
for i, row in coursera_reviews.iterrows():
    reviewer = row['reviewers']
    course_id = row['course_id']
    if pd.isna(course_id):
        if reviewer in reviewers_dict:
            coursera_reviews.at[i, 'course_id'] = reviewers_dict[reviewer]
    else:
        reviewers_dict[reviewer] = course_id
```

```python
from wordcloud import (WordCloud, STOPWORDS)
import matplotlib.pyplot as plt

# Generar una imagen de nube de palabras
#Se une el texto en una sola cadena usando el método "join()" y se almacena en la variable "text".
text =  " ".join(review for review in coursera_courses['course_id'])
#Se genera la nube de palabras con el método "WordCloud().generate(text)".
wordcloud = WordCloud(collocations=False, width = 300,
    height = 200,
    background_color = 'black',
    stopwords = STOPWORDS).generate(text)

plt.figure(figsize = (10, 5))
# Se muestra la nube de palabras con la función "plt.imshow()".
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.savefig('wordcloud.png', dpi=600, bbox_inches='tight')

```
